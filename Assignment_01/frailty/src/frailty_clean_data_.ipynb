```python
import pandas as pd

# Fetching raw dataset
frailty_raw_data = pd.read_csv('/content/frailty/raw_data/frailty_raw_data.csv')
```

```python
# Printing raw data
frailty_raw_data
```

```python
# Preprocessing the data for analysis

# Renaming columns to include units of measure for better clarity
frailty_raw_data = frailty_raw_data.rename(columns={
    'Height': 'Height(in)',
    'Weight': 'Weight(lb)',
    'Age': 'Age',
    'Grip_strength': 'Grip_strength(kg)',
    'Frailty': 'Frailty'
})

# Mapping categorical values in the 'Frailty' column to numerical values
# 'Y' (Yes) is mapped to 0, and 'N' (No) is mapped to 1
frailty_raw_data['Frailty'] = frailty_raw_data['Frailty'].map({'Y': 0, 'N': 1})
```

```python
# Removing rows with missing values from the dataset
frailty_clean_data = frailty_raw_data.dropna()

# Displaying the cleaned dataset
frailty_clean_data
```

```python
# Saving the cleaned dataset to a CSV file
frailty_clean_data.to_csv('/content/frailty/clean_data/frailty_clean_data.csv', index=False)
```
